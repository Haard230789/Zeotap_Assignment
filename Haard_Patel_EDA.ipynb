```python
cust_missing = cust_df.isnull().sum()
prod_missing = prod_df.isnull().sum()
trans_missing = trans_df.isnull().sum()

cust_duplicates = cust_df.duplicated().sum()
prod_duplicates = prod_df.duplicated().sum()
trans_duplicates = trans_df.duplicated().sum()

cust_missing, prod_missing, trans_missing, cust_duplicates, prod_duplicates, trans_duplicates
```

```python
region_distribution = cust_df['Region'].value_counts()

cust_df['SignupDate'] = pd.to_datetime(cust_df['SignupDate'])
signup_trends = cust_df['SignupDate'].dt.to_period('M').value_counts().sort_index()

signup_trends.plot(kind='line', title='Customer Signup Trends Over Time', figsize=(10, 5))
```

```python
category_distribution = prod_df['Category'].value_counts()

price_summary = prod_df['Price'].describe()

sns.histplot(prod_df['Price'], kde=True)
plt.title("Price Distribution")
```

```python
trans_df['TransactionDate'] = pd.to_datetime(trans_df['TransactionDate'])
trans_df['Month'] = trans_df['TransactionDate'].dt.to_period('M')
monthly_revenue = trans_df.groupby('Month')['TotalValue'].sum()

monthly_revenue.plot(kind='line', title='Monthly Revenue Trends', figsize=(10, 5))

most_sold_products = trans_df.groupby('ProductID')['Quantity'].sum().sort_values(ascending=False)
```

```python
most_sold_products
```

```python
trans_prod_df = trans_df.merge(prod_df, on='ProductID', how='inner')

revenue_by_category = trans_prod_df.groupby('Category')['TotalValue'].sum().sort_values(ascending=False)

plt.figure(figsize=(8, 6))
sns.barplot(x=revenue_by_category.index, y=revenue_by_category.values, palette="viridis")
plt.title("Revenue by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
top_customers = trans_df.groupby('CustomerID')['TotalValue'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(8, 6))
sns.barplot(x=top_customers.index.astype(str), y=top_customers.values, palette="magma")
plt.title("Top 10 Customers by Total Spend")
plt.xlabel("Customer ID")
plt.ylabel("Total Spend")
plt.xticks(rotation=45)
plt.show()
```

```python
trans_df['TransactionDate'] = pd.to_datetime(trans_df['TransactionDate'])
trans_df['Month'] = trans_df['TransactionDate'].dt.to_period('M')

transactions_by_month = trans_df.groupby('Month')['TransactionID'].count()

plt.figure(figsize=(10, 6))
transactions_by_month.plot(kind='line', marker='o', color='green')
plt.title("Transaction Frequency by Month")
plt.xlabel("Month")
plt.ylabel("Number of Transactions")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
recent_purchase = trans_df.groupby('CustomerID')['TransactionDate'].max()
recent_purchase_days = (pd.Timestamp.now() - recent_purchase).dt.days

plt.figure(figsize=(8, 6))
sns.histplot(recent_purchase_days, kde=True, bins=20, color='skyblue')
plt.title("Customer Recency Distribution")
plt.xlabel("Days Since Last Purchase")
plt.ylabel("Number of Customers")
plt.show()
```

```python
avg_transaction_value = trans_df.groupby('Month')['TotalValue'].mean()

plt.figure(figsize=(10, 6))
avg_transaction_value.plot(kind='line', marker='o', color='purple')
plt.title("Average Transaction Value by Month")
plt.xlabel("Month")
plt.ylabel("Average Transaction Value")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
merged_df = trans_df.merge(cust_df, on='CustomerID', how='inner')

merged_df = merged_df.merge(prod_df, on='ProductID', how='inner')

merged_df.head()
```

```python
merged_df.describe()
```

```python
revenue_by_location = merged_df.groupby('Region')['TotalValue'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=revenue_by_location.index, y=revenue_by_location.values, palette="coolwarm")
plt.title("Revenue by Customer Location")
plt.xlabel("Location")
plt.ylabel("Total Revenue")
plt.xticks(rotation=45)
plt.show()
```

```python
# Get the top 5 locations by revenue
top_locations = revenue_by_location.head(5).index

# Filter data for top locations
top_location_data = merged_df[merged_df['Region'].isin(top_locations)]

# Group by category within these locations
category_by_location = top_location_data.groupby(['Region', 'Category'])['TotalValue'].sum().unstack()

# Plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(category_by_location, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Revenue by Category in Top Locations")
plt.xlabel("Category")
plt.ylabel("Location")
plt.show()
```

```python
# Cross-tabulation of categories
category_affinity = pd.crosstab(merged_df['Category'], merged_df['Category'])

# Generate a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(category_affinity, annot=True, fmt="d", cmap="coolwarm")
plt.title("Product Category Affinity")
plt.xlabel("Category")
plt.ylabel("Category")
plt.show()
```
